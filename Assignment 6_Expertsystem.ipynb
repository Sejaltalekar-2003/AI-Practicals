```python
def input_patient_info():
    patient_info = {}
    patient_info['name'] = input("Patient's name: ")
    patient_info['age'] = int(input("Patient's age: "))
    patient_info['gender'] = input("Patient's gender: ")
    patient_info['weight'] = float(input("Patient's weight (in kg): "))
    patient_info['mobile_number'] = input("Patient's mobile number: ")
    return patient_info


def assessment(symptoms):
    if 'breathing' in symptoms or 'bleeding' in symptoms:
        return "Needs immediate attention."
    elif 'severe pain' in symptoms or 'loss of consciousness' in symptoms:
        return "Needs urgent care."
    elif 'moderate pain' in symptoms or 'vomiting' in symptoms:
        return "Requires semi-urgent care."
    else:
        return "Can wait for some time"


def input_symptoms():
    symptoms = input("Symptoms (comma-separated): ").split(',')
    return symptoms


def suggest_medications(symptoms):
    medications = []

    if 'breathing' in symptoms:
        medications.append("Bronchodilators")
    if 'bleeding' in symptoms:
        medications.append("Coagulants")
    if 'severe pain' in symptoms:
        medications.append("Painkillers")
    if 'vomiting' in symptoms:
        medications.append("Anti-nausea medication")
    if 'fever' in symptoms:
        medications.append("Fever reducers")
    if 'headache' in symptoms:
        medications.append("Headache relief")

    return medications


def send_notification(patient_info, priority, medications):
    message = f"Priority: {priority}\nSuggested medications: {', '.join(medications)}"
    print("Notification sent to", patient_info['mobile_number'])
    print("Message:", message)


def main():
    patient_info = input_patient_info()
    symptoms = input_symptoms()
    priority = assessment(symptoms)
    print("Priority:", priority)

    medications = suggest_medications(symptoms)
    if medications:
        print("Suggested medications:", ", ".join(medications))
    else:
        print("No specific medications suggested for these symptoms.")

    send_notification(patient_info, priority, medications)


main()
```
